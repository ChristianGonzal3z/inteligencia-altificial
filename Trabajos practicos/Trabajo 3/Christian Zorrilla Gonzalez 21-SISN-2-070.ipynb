1. Utilizando una celda de Python, crea un árbol ternario (con 3 descendientes) que reciba números y los muestre en orden ascendente, es decir, de menor a mayor. Luego, escribe un código que muestre estos números en pantalla. Asegúrate de que, al mostrar los números, se incluyan aquellos que se insertaron repetidos. Por ejemplo, si se insertaron los números 40, 2, 6, 3 y 40, el resultado debe ser: 2, 3, 6, 40, 40.

````python
class ArbolTernario:
    def __init__(self, valor):
        self.valor = valor
        self.izquierda = None
        self.medio = None
        self.derecha = None

    def insertar(self, valor):
        if valor < self.valor:
            if self.izquierda is None:
                self.izquierda = ArbolTernario(valor)
            else:
                self.izquierda.insertar(valor)
        elif valor > self.valor:
            if self.derecha is None:
                self.derecha = ArbolTernario(valor)
            else:
                self.derecha.insertar(valor)
        else:
            if self.medio is None:
                self.medio = ArbolTernario(valor)
            else:
                self.medio.insertar(valor)

    def recorrido_ordenado(self):
        resultado = []
        if self.izquierda:
            resultado.extend(self.izquierda.recorrido_ordenado())
        resultado.append(self.valor)
        if self.medio:
            resultado.extend(self.medio.recorrido_ordenado())
        if self.derecha:
            resultado.extend(self.derecha.recorrido_ordenado())
        return resultado

arbol = ArbolTernario(40)
arbol.insertar(2)
arbol.insertar(6)
arbol.insertar(3)
arbol.insertar(40)

numeros = arbol.recorrido_ordenado()

for numero in numeros:
    print(numero)
````
