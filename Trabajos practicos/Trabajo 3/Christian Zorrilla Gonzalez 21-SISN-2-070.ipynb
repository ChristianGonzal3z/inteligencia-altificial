1. Utilizando una celda de Python, crea un árbol ternario (con 3 descendientes) que reciba números y los muestre en orden ascendente, es decir, de menor a mayor. Luego, escribe un código que muestre estos números en pantalla. Asegúrate de que, al mostrar los números, se incluyan aquellos que se insertaron repetidos. Por ejemplo, si se insertaron los números 40, 2, 6, 3 y 40, el resultado debe ser: 2, 3, 6, 40, 40.

````python
class ArbolTernario:
    def __init__(self, valor):
        self.valor = valor
        self.menor = None
        self.igual = None
        self.mayor = None

    def insertar(self, valor):
        if valor < self.valor:
            if self.menor is None:
                self.menor = ArbolTernario(valor)
            else:
                self.menor.insertar(valor)
        elif valor > self.valor:
            if self.mayor is None:
                self.mayor = ArbolTernario(valor)
            else:
                self.mayor.insertar(valor)
        else:
            if self.igual is None:
                self.igual = ArbolTernario(valor)
            else:
                self.igual.insertar(valor)

    def recorrer(self):
        resultado = []
        if self.menor is not None:
            resultado.extend(self.menor.recorrer())
        if self.igual is not None:
            resultado.extend(self.igual.recorrer())
        resultado.append(self.valor)
        if self.mayor is not None:
            resultado.extend(self.mayor.recorrer())
        return resultado


arbol = None

while True:
    numero = input("Ingrese un número (o 'q' para salir): ")
    if numero == 'q':
        break
    try:
        numero = int(numero)
        if arbol is None:
            arbol = ArbolTernario(numero)
        else:
            arbol.insertar(numero)
    except ValueError:
        print("Entrada inválida. Por favor, ingrese un número entero.")

if arbol is not None:
    numeros = arbol.recorrer()
    for numero in numeros:
        print(numero, end=", ")
````
