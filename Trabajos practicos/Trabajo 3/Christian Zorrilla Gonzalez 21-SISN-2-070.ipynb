1. Utilizando una celda de Python, crea un árbol ternario (con 3 descendientes) que reciba números y los muestre en orden ascendente, es decir, de menor a mayor. Luego, escribe un código que muestre estos números en pantalla. Asegúrate de que, al mostrar los números, se incluyan aquellos que se insertaron repetidos. Por ejemplo, si se insertaron los números 40, 2, 6, 3 y 40, el resultado debe ser: 2, 3, 6, 40, 40.

````python
class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.menor = None
        self.igual = None
        self.mayor = None

def insertar_nodo(raiz, valor):
    if raiz is None:
        return Nodo(valor)
    
    if valor < raiz.valor:
        raiz.menor = insertar_nodo(raiz.menor, valor)
    elif valor == raiz.valor:
        raiz.igual = insertar_nodo(raiz.igual, valor)
    else:
        raiz.mayor = insertar_nodo(raiz.mayor, valor)
    
    return raiz

def mostrar_arbol(raiz):
    if raiz is not None:
        mostrar_arbol(raiz.menor)
        print(raiz.valor)
        mostrar_arbol(raiz.igual)
        mostrar_arbol(raiz.mayor)

numeros = input("Ingresa los numeros separados por espacios: ").split()

arbol = None
for numero in numeros:
    arbol = insertar_nodo(arbol, int(numero))

mostrar_arbol(arbol)
````
